```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols),
        ('num', StandardScaler(), numerical_cols)
    ],
    remainder='drop'
)
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=500, class_weight='balanced'),
    "Random Forest": RandomForestClassifier(random_state=42, class_weight='balanced'),
    "SVM": SVC(probability=True, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    "KNN": KNeighborsClassifier()
}

results = {}
detail_reports = {}

for name, model in models.items():
    pipe = Pipeline(steps=[('preproc', preprocessor), ('model', model)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    try:
        y_proba = pipe.predict_proba(X_test)[:,1]
        auc = roc_auc_score(y_test, y_proba)
    except Exception:
        auc = None
    results[name] = {'accuracy': acc, 'roc_auc': auc}
    detail_reports[name] = {
        'classification_report': classification_report(y_test, y_pred, output_dict=True),
        'confusion_matrix': confusion_matrix(y_test, y_pred)
    }
    print(f"--- {name} ---")
    print("Accuracy:", acc)
    if auc is not None:
        print("ROC AUC:", round(auc,3))
    print(classification_report(y_test, y_pred))
    print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
    print()
```

```python
import pandas as pd
import numpy as np
import sklearn

def make_row_from_sample(sample_dict, X_columns, X_dtypes, fill_value=np.nan):
    """
    sample_dict: dict of some feature-name: value pairs (names should be a subset of X_columns)
    X_columns: list-like columns from original X (in the same order used for training)
    X_dtypes: a pandas Series or dict mapping column -> dtype from original X (X.dtypes)
    """
    row = pd.DataFrame([sample_dict])
    for c in X_columns:
        if c not in row.columns:
            row[c] = fill_value
    row = row.loc[:, X_columns]
    for c in X_columns:
        try:
            dtype = X_dtypes[c]
            if pd.api.types.is_integer_dtype(dtype):
                row[c] = pd.to_numeric(row[c], errors='coerce').astype('Int64')  # nullable int
            elif pd.api.types.is_float_dtype(dtype):
                row[c] = pd.to_numeric(row[c], errors='coerce').astype(float)
            elif pd.api.types.is_bool_dtype(dtype):
                row[c] = row[c].astype('boolean')
            else:
                row[c] = row[c].astype(object)
        except Exception:
            pass
    return row

original_cols = X.columns.tolist()
original_dtypes = X.dtypes                

sample = {
    'age': 58,
    'resting_blood_pressure': 140,
    'serum_cholesterol_mg_per_dl': 240,
    'max_heart_rate_achieved': 150,
    'oldpeak_eq_st_depression': 1.2,
    'chest_pain_type': 3,
    'slope_of_peak_exercise_st_segment': 2,
    'exercise_induced_angina': 0,
    'fasting_blood_sugar_gt_120_mg_per_dl': 0,
    'resting_ekg_results': 2,
    'sex': 1,
    'num_major_vessels': 3,
}

row = make_row_from_sample(sample, original_cols, original_dtypes, fill_value=np.nan)
print("Row columns:", row.columns.tolist())
print(row)


proba = final_pipeline.predict_proba(row)[:,1]  
pred = final_pipeline.predict(row)
print("pred:", pred, "proba:", proba)
```
